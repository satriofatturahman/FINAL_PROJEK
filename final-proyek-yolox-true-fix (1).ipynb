```python
from ultralytics import YOLO
# Load a pre-trained model 
model = YOLO('yolov8x-seg.pt')

# Train the model 
results = model.train(data='/kaggle/working/data.yaml', epochs=50, batch = 16, imgsz=(736,544), device = [0,1] , workers = 8, pretrained = True, optimizer = 'SGD', lr0 = 0.005, momentum = 0.9) # argumen tinggal tambah
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 segmentation model
model = YOLO("/kaggle/working/runs/segment/train3/weights/best.pt")
```

```python
from PIL import Image 
  
# get image 
img_path = "/kaggle/input/new-beachdataset-yolov8-080924/New-Beachdata/Test/images/003143.jpg"
img_temp = Image.open(img_path) 
  
# get width and height 
width = img_temp.width 
height = img_temp.height 

results = model.predict(img_path, save=True, imgsz=(height, width), conf=0.3, show_labels = False, show_conf = False, show_boxes = False)
```
